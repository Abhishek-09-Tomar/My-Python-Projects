## Train-Test Splitting

```python
import numpy as np
def split_train_test(data, test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    print(shuffled)
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size = 0.2, random_state = 42)
print(f"Rows in train set: {len(train_set)}\n Rows in test_set:{len(test_set)}")
```

## Apply imputer for fitting the missing values of 'RM' through median 

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
imputer.fit(df)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(df)
new_df = pd.DataFrame(X, columns=df.columns)
new_df.describe()
```

## As you observe in the data frame our RM value is missing and its value is (501) before fitting and after fitting the missing values of RM we get overall values (506)

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler())
])
```

```python
new_df_tr = my_pipeline.fit_transform(df)
new_df_tr
```

## Selection of a Desired model for Evergreens Estates

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model = DecisionTreeRegressor()
# model = LinearRegression()
model = RandomForestRegressor()
model.fit(new_df_tr, df_labels)
```

## Evaluating The Model

```python
from sklearn.metrics import mean_squared_error
df_predictions = model.predict(new_df_tr)
mse = mean_squared_error(df_labels, df_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model , new_df_tr, df_labels, scoring='neg_mean_squared_error', cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("scores :", scores)
    print("Mean :", scores.mean())
    print("Standard_deviation :", scores.std())
```

```python
print_scores(rmse_scores)
```
